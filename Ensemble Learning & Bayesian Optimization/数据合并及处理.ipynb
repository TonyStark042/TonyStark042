```python
import pandas as pd
import numpy as np
```

```python
account_list = ['普华永道', '瑞华', '德勤', '立信', '安永华明', '毕马威', '天健', '信永中和', '天职国际', '致同']
```

```python
new_data = pd.read_excel('增加指标.xlsx')
origin_data = pd.read_excel('原始数据.xlsx')
```

```python
new_data.drop(labels=['Unnamed: 2','Unnamed: 3', '违约日期'], axis = 1,inplace=True)
new_data['独立董事比例'] = new_data['独立董事']/new_data['董事']
new_data.drop(labels=['独立董事','董事'], axis = 1, inplace=True)
```

```python
new_data_fill0 = new_data.fillna(0)
```

```python
del_row = []
for i in range(567):
    count_0 = 0
    for j in range(9):
        if new_data_fill0.iloc[i][j] == 0:
            count_0 += 1
    if count_0 >= 3:
        del_row.append(i)
```

```python
total_data = pd.concat([origin_data,new_data], axis = 1)
total_data.insert(39, '是否违约', total_data.pop('是否违约')) 
```

```python
total_data
```

```python
total_data.drop(labels=del_row, axis = 0, inplace=True)
total_data.reset_index(drop=True, inplace=True)
```

```python
for i in range(404):
    for j in account_list:
        if j in total_data.loc[i, '会计']:
            total_data.loc[i, '会计'] = 1
            break
    else:
        total_data.loc[i, '会计'] = 0
```

```python
total_data['机构持股'].fillna(total_data['机构持股'].mean(),inplace=True)
total_data['股权集中度'].fillna(total_data['股权集中度'].mean(),inplace=True)
```

```python
total_data.to_excel('total_data.xlsx')
```
