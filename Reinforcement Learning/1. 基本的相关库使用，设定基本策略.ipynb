### MountainCar-v0 例子

```python
env = gym.make('MountainCar-v0', render_mode = 'human')
print('观测空间 = {}'. format(env.observation_space))
print('动作空间 = {}'. format(env.action_space)) 
print('观测范围 = {} ~ {}'. format(env.observation_space.low,
env.observation_space.high))
print('动作数 = {}'. format(env.action_space.n))

# 观测空间表示，观测的状态有两个值，类型均为float，以及两个值的范围。
# 动作空间表示有离散的三个值。
```

```python
# 实现智能体来控制小车移动

class SimpleAgent:
    def __init__(self, env):
        pass

    def decide(self, observation):                # 决策
        position, velocity = observation          # obsevation的两维参数分别表示位置和动作。
        lb = min(-0.09 * (position + 0.25) ** 2 + 0.03, 0.3 * (position + 0.9) ** 4 - 0.008)
        ub = -0.07 * (position + 0.38) ** 2 + 0.07
        if lb < velocity < ub:                    # 设定速度的上下界
            action = 2
        else:
            action = 0
        return action         # 返回动作
    
    def learn(self, *args):   # 不确定传入参数的数量，*args以元组的形式传递参数。
        pass
        
agent = SimpleAgent(env)      # Agent是由策略类创造的
```

```python
# 定义智能体与环境的交互

def play(env, agent, train=False):
    episode_reward = 0.                        # 记录回合总奖励，初始值为0
    observation, _ = env.reset(seed=3)            # 随机性的重置游戏新起始状态，开始新的回合
                                               # reset的返回值包括了obervation和info，这里记得要只取observation
    
    while True:                                # 不断循环，直到回合结束
        action = agent.decide(observation)     # 根据obervation信息执行动作
        next_observation, reward, done, _, _ = env.step(action) # 执行动作
        episode_reward += reward               # 收集回合奖励
        if train:                              # 判断是否训练智能体
            agent.learn(observation, action, reward, done) # 学习
        if done: # 回合结束，跳出循环
            break
        observation = next_observation
    
    return episode_reward # 返回回合总奖励
```

```python
# 运行程序

env = gym.make('MountainCar-v0', render_mode = 'human')
episode_reward = play(env, agent)
print(f'总回合奖励 = {episode_reward}')

env.close() # 关闭图形界面
```

```python
# 测得10回合的平均奖励
import numpy as np

env = gym.make('MountainCar-v0', render_mode = 'human')
episode_rewards = [play(env, agent) for _ in range(10)]
print(f'平均回合奖励 = {np.mean(episode_rewards)}')

env.close() # 关闭图形界面
```
